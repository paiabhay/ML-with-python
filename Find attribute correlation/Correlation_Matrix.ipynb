```python
# Standard plotly imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read dataset
data = pd.read_csv('datasets/diabetes.csv')
```

# Print correlation matrix
## corr()
This function helps to compute pairwise correlation of columns. <br>
Link: <a href="https://bit.ly/2FnSOJI">sklearn Correlation</a>

```python
# Compute correlationbetween columns
corr = data.corr()
corr.style.background_gradient(cmap='copper')
```

```python
# Plot correlation graph
def correlation_matrix_plot(data):
    plt.figure(figsize=(8,8))
    plt.matshow(data.corr(), cmap=plt.get_cmap('copper'), fignum=1)
    plt.xticks(ticks=range(len(data.columns)), labels=data.columns, rotation='vertical')
    plt.yticks(range(len(data.columns)), data.columns)
    plt.colorbar()
    plt.ylabel('Correlation Plot').set_fontsize(40)
    plt.show()
```

```python
correlation_matrix_plot(data)
```
