# Plotting correlation between attributes
Here we will try to plot co-relation between attributes.<br>
Dataset Link: https://goo.gl/1Awfqk

```
# Helping libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Read dataset
df = pd.read_csv('indian_liver_patient.csv')
```

## corr()
Used to compute pairwise corelation between attributes of dataset.<br>
Link: https://goo.gl/zdDMhX

```
correlation = df.corr(method='spearman')
print(correlation)
```

## Plot correlation

```
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlation, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(df.columns), 1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
